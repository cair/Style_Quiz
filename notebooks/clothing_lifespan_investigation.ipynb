```python
from tqdm import tqdm
tqdm.pandas()

def get_name_status(outfit_name):
    split = outfit_name.split("-")
    if len(split) > 1:
        return split[0]
    return ""

outfits_df["name_status"] = outfits_df["name"].progress_apply(get_name_status)
```

```python
#outfits_df[outfits_df["isDeleted"] == 1].sort_values(["id", "meta.validTo"], ascending=False).groupby("id").first().reset_index()
eventually_deleted_ids = outfits_df[outfits_df["isDeleted"] == 1]["id"].unique()
eventually_deleted_outfits = outfits_df[outfits_df["id"].isin(eventually_deleted_ids)].sort_values(["id", "meta.validTo"], ascending=False)
```

```python
deleted_outfits_creation_time = eventually_deleted_outfits.groupby("id")["timeCreated"].min()
deleted_outfits_deletion_time = eventually_deleted_outfits[eventually_deleted_outfits["isDeleted"] == 1].groupby("id")["timeUpdated"].min()
eventually_deleted_outfits["originally_created"] = eventually_deleted_outfits["id"].map(deleted_outfits_creation_time)
eventually_deleted_outfits["originally_deleted"] = eventually_deleted_outfits["id"].map(deleted_outfits_deletion_time)
eventually_deleted_outfits["lifespan"] = pd.to_datetime(eventually_deleted_outfits["originally_deleted"]) - pd.to_datetime(eventually_deleted_outfits["originally_created"])
lifespan_df = eventually_deleted_outfits.groupby("id").first().reset_index()
lifespan_df = lifespan_df[lifespan_df["lifespan"] > pd.Timedelta(days=60)].copy()
```

```python
import numpy as np

def get_outfit_category(tag_categories, tags, category):
    tag_categories, tags = np.array(tag_categories), np.array(tags)
    category_indexes = np.where(tag_categories == category)[0]
    if len(category_indexes) == 0:
        return []
    cat_tags = tags[category_indexes]
    return cat_tags

lifespan_df["brand"] = lifespan_df.apply(lambda x: get_outfit_category(x["tag_categories"], x["outfit_tags"], "Brand"), axis=1)
lifespan_df["material"] = lifespan_df.apply(lambda x: get_outfit_category(x["tag_categories"], x["outfit_tags"], "Material"), axis=1)
lifespan_df["size"] = lifespan_df.apply(lambda x: get_outfit_category(x["tag_categories"], x["outfit_tags"], "Size"), axis=1)
lifespan_df["category"] = lifespan_df.apply(lambda x: get_outfit_category(x["tag_categories"], x["outfit_tags"], "Category"), axis=1)
lifespan_df["occasion"] = lifespan_df.apply(lambda x: get_outfit_category(x["tag_categories"], x["outfit_tags"], "Occasion"), axis=1)
```

```python
all_brand_classes = lifespan_df["brand"].explode()
mlb_brand_classes = all_brand_classes.value_counts()[all_brand_classes.value_counts() > 50].index.to_list()

all_material_classes = lifespan_df["material"].explode()
mlb_material_classes = all_material_classes.value_counts()[all_material_classes.value_counts() > 50].index.to_list()

all_category_classes = lifespan_df["category"].explode()
mlb_category_classes = all_category_classes.value_counts()[all_category_classes.value_counts() > 50].index.to_list()

all_occasion_classes = lifespan_df["occasion"].explode()
mlb_occasion_classes = all_occasion_classes.value_counts()[all_occasion_classes.value_counts() > 0].index.to_list()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
OTHER_PLACEHOLDER_NAME = "Other"

def remove_rarely_used_labels(labels, used_labels):
    return set([label if label in used_labels else OTHER_PLACEHOLDER_NAME for label in labels])

mlb_brand_classes.append(OTHER_PLACEHOLDER_NAME)
brand_mlb = MultiLabelBinarizer(classes=mlb_brand_classes)
lifespan_df["culled_brand"] = lifespan_df["brand"].apply(lambda x: remove_rarely_used_labels(x, mlb_brand_classes))
brand_df = pd.DataFrame(brand_mlb.fit_transform(lifespan_df["culled_brand"]), columns=["brand_" + mlb_class for mlb_class in brand_mlb.classes_],  index=lifespan_df.index)

mlb_material_classes.append(OTHER_PLACEHOLDER_NAME)
material_mlb = MultiLabelBinarizer(classes=mlb_material_classes)
lifespan_df["culled_material"] = lifespan_df["material"].apply(lambda x: remove_rarely_used_labels(x, mlb_material_classes))
material_df = pd.DataFrame(material_mlb.fit_transform(lifespan_df["culled_material"]), columns=["material_" + mlb_class for mlb_class in material_mlb.classes_],  index=lifespan_df.index)

mlb_category_classes.append(OTHER_PLACEHOLDER_NAME)
category_mlb = MultiLabelBinarizer(classes=mlb_category_classes)
lifespan_df["culled_category"] = lifespan_df["category"].apply(lambda x: remove_rarely_used_labels(x, mlb_category_classes))
category_df = pd.DataFrame(category_mlb.fit_transform(lifespan_df["culled_category"]), columns=["category_" + mlb_class for mlb_class in category_mlb.classes_],  index=lifespan_df.index)

#mlb_occasion_classes.append(OTHER_PLACEHOLDER_NAME)
occasion_mlb = MultiLabelBinarizer(classes=mlb_occasion_classes)
#lifespan_df["culled_occasion"] = lifespan_df["occasion"].apply(lambda x: remove_rarely_used_labels(x, mlb_occasion_classes))
occasion_df = pd.DataFrame(occasion_mlb.fit_transform(lifespan_df["occasion"]), columns=["occasion_" + mlb_class for mlb_class in occasion_mlb.classes_],  index=lifespan_df.index)
```

```python
lifespan_df = pd.concat([lifespan_df, brand_df, material_df, category_df, occasion_df], axis=1)
lifespan_df["lifespan_days"] = lifespan_df["lifespan"].dt.days
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split


from sklearn.preprocessing import MinMaxScaler

brand_columns = [col for col in lifespan_df.columns if col.startswith("brand_")]
material_columns = [col for col in lifespan_df.columns if col.startswith("material_")]
category_columns = [col for col in lifespan_df.columns if col.startswith("category_")]
occasion_columns = [col for col in lifespan_df.columns if col.startswith("occasion_")]
LR_COLUMNS = ["retailPrice", "rentals_count"] + brand_columns + material_columns + category_columns + occasion_columns
PREDICTION_COLUMN = "lifespan_days"

train_df, test_df = train_test_split(lifespan_df, test_size=0.2, random_state=42)
scaler = MinMaxScaler()
train_df[LR_COLUMNS] = scaler.fit_transform(train_df[LR_COLUMNS])
test_df[LR_COLUMNS] = scaler.transform(test_df[LR_COLUMNS])

train_x, train_y = train_df[LR_COLUMNS].fillna(0).values, train_df[PREDICTION_COLUMN].values.ravel()

model = LinearRegression()
#model = LogisticRegression(max_iter=10**3)
#model = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=SEED)
model.fit(train_x, train_y)
```

```python
test_x = test_df[LR_COLUMNS].fillna(0).values

test_df["predicted_lifespan"] = model.predict(test_x)
test_df["prediction_error"] = abs(test_df["predicted_lifespan"] - test_df["lifespan_days"])
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import DataLoader, TensorDataset

#train_df, test_df = train_test_split(lifespan_df, test_size=0.2, random_state=42)
scaler = MinMaxScaler()
train_df[LR_COLUMNS] = scaler.fit_transform(train_df[LR_COLUMNS])
test_df[LR_COLUMNS] = scaler.transform(test_df[LR_COLUMNS])
scaler_y = MinMaxScaler()
train_df[PREDICTION_COLUMN] = scaler_y.fit_transform(train_df[[PREDICTION_COLUMN]])
test_df[PREDICTION_COLUMN] = scaler_y.transform(test_df[[PREDICTION_COLUMN]])

train_x, train_y = train_df[LR_COLUMNS].fillna(0).values, train_df[PREDICTION_COLUMN].values
test_x, test_y = test_df[LR_COLUMNS].fillna(0).values, test_df[PREDICTION_COLUMN].values

# Convert numpy arrays to PyTorch tensors
train_x = torch.tensor(train_x, dtype=torch.float32)
train_y = torch.tensor(train_y, dtype=torch.float32).view(-1, 1)  # Reshape to [n_samples, 1]
test_x = torch.tensor(test_x, dtype=torch.float32)
test_y = torch.tensor(test_y, dtype=torch.float32).view(-1, 1)

# Neural Network
class Net(nn.Module):
    def __init__(self, input_size):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(input_size, 32)  
        self.fc2 = nn.Linear(32, 64)
        self.fc3 = nn.Linear(64, 1) 

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Initialize network
input_size = train_x.shape[1]
model = Net(input_size)

# Loss and Optimizer
criterion = nn.MSELoss()  # Mean Squared Error Loss for regression tasks
optimizer = optim.Adam(model.parameters(), lr=0.01)  # Adam optimizer

# Training Loop
epochs = 100
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(train_x)
    loss = criterion(outputs, train_y)
    loss.backward()
    optimizer.step()
    
    model.eval()
    with torch.no_grad():
        test_outputs = model(test_x)
        test_loss = criterion(test_outputs, test_y)
    
    if epoch % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}, Test Loss: {test_loss.item():.4f}')

# You might want to add more to the evaluation, such as saving the model, or more detailed performance metrics.
```
