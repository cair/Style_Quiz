```python
from FREja_API.constants import *
import pandas as pd
import numpy as np
import os

outfits_df = pd.read_pickle(os.path.join(CBM_SAVE_PATH, OUTFITS_PATH), compression=COMPRESSION_TYPE)

def get_outfit_category(outfit_row, category, take_first=True):
    category_index = np.where(outfit_row["tag_categories"] == category)[0]
    if len(category_index) == 0:
        return None
    if take_first:
        return outfit_row["outfit_tags"][category_index][0]
    return outfit_row["outfit_tags"][category_index]

outfits_df["category"] = outfits_df.apply(lambda x: get_outfit_category(x, "Category", take_first=True), axis=1)
outfits_df["all_categories"] = outfits_df.apply(lambda x: get_outfit_category(x, "Category", take_first=False), axis=1)

outfits_df.head()
```

```python
als_outfits_df = pd.read_pickle(os.path.join(ALS_SAVE_PATH, OUTFITS_PATH), compression=COMPRESSION_TYPE)
als_predictions_df = pd.read_pickle(os.path.join(ALS_SAVE_PATH, PREDICTIONS_PATH), compression=COMPRESSION_TYPE).reset_index()
als_dict = {}
for index, row in als_predictions_df.iterrows():
    key = row["input_outfits"]
    values = row.drop(["input_outfits"]).values
    als_dict[key] = values
```

```python
outfits_df["ALS_embedding"] = outfits_df["id"].apply(lambda x: als_dict[x] if x in als_dict else None)
outfits_df["ALS_embedding"] = outfits_df["ALS_embedding"].apply(lambda x: np.array(x))

def create_outfit_representation(outfit_row):
    combined_representation = np.concatenate([outfit_row["tag_encoding"], outfit_row["embedding"], outfit_row["ALS_embedding"]])
    return combined_representation

outfits_df["outfit_representation"] = outfits_df.apply(create_outfit_representation, axis=1)
print(np.stack(outfits_df["outfit_representation"].values).shape)
```

```python
embeddings_values = np.stack(outfits_df["embedding"].values)
pca_embeddings_values = np.stack(outfits_df["pca_embedding"].values)
outfit_representations = np.stack(outfits_df["outfit_representation"].values)
weighted_tag_embeddings = np.stack(outfits_df["weighted_tag_encoding"].values)

print(embeddings_values.shape, pca_embeddings_values.shape, outfit_representations.shape, weighted_tag_embeddings.shape)
```

```python
import pandas as pd
import numpy as np
import torch
from torch.utils.data import DataLoader, TensorDataset
from torch.nn.functional import normalize

embeddings = torch.tensor(outfit_representations.astype(np.float32).tolist())
embeddings = normalize(embeddings, dim=1)

dataset = TensorDataset(embeddings, embeddings)  # We use embeddings as both input and target
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
from torch import nn

class Autoencoder(nn.Module):
    def __init__(self, encoding_dim):
        super(Autoencoder, self).__init__()
        # Encoder
        self.encoder = nn.Sequential(
            nn.Linear(embeddings.shape[1], embeddings.shape[1] // 2),
            nn.ReLU(True),
            nn.Linear(embeddings.shape[1] // 2, encoding_dim),
            nn.LeakyReLU(0.1),
        )
        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(encoding_dim, embeddings.shape[1] // 2),
            nn.ReLU(True),
            nn.Linear(embeddings.shape[1] // 2, embeddings.shape[1]),
            #nn.ReLU(True),
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

encoding_dim = 50  # This is a hyperparameter you can tune
autoencoder = Autoencoder(encoding_dim).to(device)

optimizer = torch.optim.Adam(autoencoder.parameters(), lr=1e-3)
criterion = nn.MSELoss()

# Training loop
num_epochs = 20  # This is another hyperparameter to tune
for epoch in range(num_epochs):
    for data in dataloader:
        inputs, targets = data
        inputs, targets = inputs.to(device), targets.to(device)
        # Forward pass
        outputs = autoencoder(inputs)
        loss = criterion(outputs, targets)

        # Backward pass and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.8f}')
```

```python
from tqdm.notebook import tqdm

tqdm.pandas()

def apply_autoencoder(embedding):
    embedding = torch.tensor(embedding.astype(np.float32).tolist()).to(device)
    embedding = normalize(embedding.reshape(1, -1), dim=1)
    embedding = autoencoder.encoder(embedding)
    return embedding.cpu().detach().numpy()[0]

with torch.no_grad():
    outfits_df["autoencoder_embedding"] = outfits_df["outfit_representation"].progress_apply(apply_autoencoder)
```

```python
CONTRASTIVE_TAG_CATEGORIES = ["Seasons", "Occasion", "Material", "Color", "Details"]

def get_contrastive_outfit_tags(outfits_row):
    tags = outfits_row["outfit_tags"]
    tag_categories = outfits_row["tag_categories"]
    contrastive_tags, contrastive_categories = [], []
    for tag, category in zip(tags, tag_categories):
        if category in CONTRASTIVE_TAG_CATEGORIES:
            contrastive_tags.append(tag)
            contrastive_categories.append(category)
    return contrastive_tags, contrastive_categories

outfits_df["contrastive_tags"], outfits_df["contrastive_categories"] = zip(*outfits_df.progress_apply(get_contrastive_outfit_tags, axis=1))
```
